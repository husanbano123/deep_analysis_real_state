# STEP 5 : Statistical Analysis
The one-sample t-test is typically used to test whether the mean of a sample is significantly different from a known or hypothesized population mean.

```python
df.describe()
```

the avg Y house price of unit area is above 37.9.

H0=mu(mean of population)<=37.9
H1=mu>37.9


```python
from scipy import stats
house_price_data=df['Y house price of unit area']

#calculate sample mean from descriptive stats
sample_mean=df['Y house price of unit area'].mean()

alpha=0.05

#perform one sample t test
t_stats,p_value=stats.ttest_1samp(house_price_data,popmean=37.5)
print("\nOne-Sample t-test Results:")
print("t-statistics:", t_stats)
print("P-value:", p_value)

#cheak for significance
if (p_value<alpha):
    print("\nRejecting the null hypo H0 ,suggesting avg house price of unit area is siginificantly more than 37.98")
else:
    print("\nFail to Reject null hypothesis H0,suggesting we do not find sufficient evidence to conclude that the mean house price is different from the hypothesized population mean. ")
```

# 3.Model Training and Evaluation:


# Train and Test the Module

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
def build_and_train_model(df,target_name,reg_fn):
    X=df.drop(target_name,axis=1)
    Y=df[target_name]
    X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
    model=reg_fn(X_train,Y_train)
    score=model.score(X_train,Y_train)
    print("\nTraining Score: ",score)
    Y_pred=model.predict(X_test)
    r_score=r2_score(Y_test,Y_pred)
    mean_absolute=mean_absolute_error(Y_test,Y_pred)
    mean_square=mean_squared_error(Y_test,Y_pred)
    print("\nTesting Score r2_score: ",r_score)
    print("Testing Score mean_absolute_error: ",mean_absolute)
    print("Testing Score mean square_error: ",mean_square)
    return {'model':model,
           'X_train':X_train,'Y_train':Y_train,
            'X_test':X_test,'Y_test':Y_test,
            'Y_pred':Y_pred
           }
```

```python
linear_reg=build_and_train_model(df,'Y house price of unit area',linear_model)
```

The MAE and MSE scores (0.3904 and 0.2897) indicate that the model's predictions are relatively close to the actual values in the testing set.

based on these evaluation metrics, the linear regression model appears to perform reasonably well, with a good balance between training and testing performance.

# 4.Model Tuning:
Hyperparameter tuning using techniques like grid search or random search, along with cross-validation, is a crucial step to optimize the performance of machine learning models. Here's how you can perform model tuning using these techniques:

Use Ridge regression cross-validation to assess the model's generalization performance and reduce overfitting.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
ridge_model = Ridge(alpha=0.9,max_iter=None)
#ridge_regression=Ridge(alpha=0.9,max_iter=None)
ridge_model.fit(X_train,Y_train)

param_grid = {'alpha': [0.1, 1.0, 10.0]}
grid_search = GridSearchCV(ridge_model, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, Y_train)
best_model = grid_search.best_estimator_
```

```python
# Example: Evaluate model performance
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

y_pred = best_model.predict(X_test)
mse = mean_squared_error(Y_test, y_pred)
mae = mean_absolute_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print("MSE:", mse)
print("MAE:", mae)
print("R-squared:", r2)
```

# 5. Model Explanation

Shapley Values: Use SHAP (SHapley Additive exPlanations) values to explain individual predictions and feature contributions. 

```python
import shap

# Initialize SHAP explainer
explainer = shap.Explainer(ridge_model, X_train)

# Calculate SHAP values for a sample of data points
shap_values = explainer(X_test)

# Visualize SHAP summary plot
shap.summary_plot(shap_values, X_test, feature_names=feature_names)
```

Positive SHAP values indicate that the feature contributes to increasing the prediction, while negative values contribute to decreasing the prediction.

Positive Impact Features:
Identify features with consistently positive SHAP values across multiple data points.
These features contribute positively to higher predictions and may be important for understanding high-value outcomes.

Negative Impact Features:
Identify features with consistently negative SHAP values.
These features contribute negatively to predictions and may be associated with lower outcomes.

Feature Importance:
Evaluate the overall importance of features based on their absolute mean SHAP values.
Features with higher absolute mean SHAP values have a stronger impact on the model's predictions.
